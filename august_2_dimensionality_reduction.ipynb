## Dimensionality Reduction 
    #### Principal Component Analysis(PCA)

- Reduce dimensionality to visualize the clusters in 2D or 3D.
- Techniques like Principal Component Analysis (PCA) or t-SNE can be used for dimensionality reduction.
- PCA:
    - `from sklearn.decomposition import PCA`: Import the PCA module.
    - `pca = PCA(n_components=2)`: Create an instance of PCA with the desired number of components.



### Summary:
- The code begins by importing necessary libraries (`pandas`, `PCA`, `seaborn`, `matplotlib`).
- A sample dataset is created using `pandas`.
- PCA is applied to reduce the dimensionality of the dataset from 3 to 2 components.
- The principal components are stored in a new DataFrame.
- A scatter plot is created to visualize the data in the new 2D space formed by the principal components.


```python
import pandas as pd
from sklearn.decomposition import PCA
! pip install seaborn
import seaborn as sns
import matplotlib.pyplot as plt
```

1. **`import pandas as pd`**: This imports the `pandas` library, which is used for data manipulation and analysis, particularly with DataFrames. The alias `pd` is commonly used to shorten the library's name in the code.

2. **`from sklearn.decomposition import PCA`**: This imports the `PCA` class from the `sklearn.decomposition` module. `PCA` (Principal Component Analysis) is a dimensionality reduction technique that transforms the features in your dataset into a smaller set of orthogonal (uncorrelated) components.

3. **`import seaborn as sns`**: This imports the `seaborn` library, which is a statistical data visualization library built on top of `matplotlib`. The alias `sns` is used to access `seaborn` functions.

4. **`import matplotlib.pyplot as plt`**: This imports `pyplot` from the `matplotlib` library, which is used for creating static, interactive, and animated visualizations in Python. The alias `plt` is used to access `pyplot` functions.


```python
data = pd.DataFrame({
    'Feature 1': [2.5, 0.5, 2.2, 1.9, 3.1, 2.3, 2, 1, 1.5, 1.1],
    'Feature 2': [2.4, 0.7, 2.9, 2.2, 3, 2.7, 1.6, 1.1, 1.6, 0.9],
    'Feature 3': [3.5, 2.4, 1.0, 1.5, 2.5, 3.6, 2.0, 4.0, 3.0, 2.0]
})
```

5. **`data = pd.DataFrame({...})`**: This line creates a `pandas` DataFrame named `data`. A DataFrame is a two-dimensional, tabular data structure with labeled axes (rows and columns). In this case, the DataFrame has three columns: 'Feature 1', 'Feature 2', and 'Feature 3', each containing a list of 10 numerical values.

    - **'Feature 1'**, **'Feature 2'**, **'Feature 3'**: These are the names of the columns in the DataFrame, representing three different features (or dimensions) of the data.
    - The lists associated with each column represent the values for that feature.


```python
data
```

```python
pca = PCA(n_components=2)
```

    - **`n_components=2`**: This parameter specifies the number of principal components to compute. Here, the data will be reduced from 3 dimensions (features) to 2 dimensions. The `PCA` algorithm will find the two principal components that capture the most variance in the data.

```python
pca
```

```python
principal_components = pca.fit_transform(data)
```

7. **`principal_components = pca.fit_transform(data)`**: This line applies PCA to the DataFrame `data` and returns the transformed data (principal components), which is stored in the variable `principal_components`.

    - **`pca.fit_transform(data)`**: The `fit_transform()` method first fits the PCA model to the data (i.e., it computes the principal components that capture the most variance), and then it transforms the original data into the new principal component space. The result is an array where each row represents a data point, and each column represents a principal component.


```python
principal_components
```

```python
principal_df = pd.DataFrame(principal_components, columns=['PC1', 'PC2'])
```

8. **`principal_df = pd.DataFrame(principal_components, columns=['PC1', 'PC2'])`**: This line creates a new DataFrame `principal_df` from the `principal_components` array. 

    - **`principal_components`**: This is the 2D array obtained from the PCA transformation, where each row is a data point and each column corresponds to a principal component.
    - **`columns=['PC1', 'PC2']`**: This argument assigns names to the columns in the new DataFrame, labeling the first principal component as 'PC1' and the second as 'PC2'.


```python
sns.scatterplot(x='PC1', y='PC2', data=principal_df)
```

9. **`sns.scatterplot(x='PC1', y='PC2', data=principal_df)`**: This line creates a scatter plot using `seaborn` to visualize the data points in the space of the two principal components.

    - **`x='PC1'`**: Specifies that the 'PC1' column of the `principal_df` DataFrame should be used for the x-axis of the scatter plot.
    - **`y='PC2'`**: Specifies that the 'PC2' column should be used for the y-axis.
    - **`data=principal_df`**: Indicates that the data for the plot is coming from the `principal_df` DataFrame.


```python
plt.title("PCA of Sample Data")
plt.show()
```

10. **`plt.title("PCA of Sample Data")`**: This line sets the title of the plot to "PCA of Sample Data" using `matplotlib`'s `plt` module.

11. **`plt.show()`**: This command displays the plot on the screen. Without this line, the plot would not be rendered in most environments.
