### Density Based Spatial Clustering Applicaiton With Noise (DBSCAN)
- robust to noise
- identify the crime - outliers




### Import the modules to use


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
```

   - `matplotlib.pyplot`: A plotting library used to create graphs and visualizations.
   - `make_moons`: A function from `sklearn.datasets` that generates a toy dataset with two interleaving half circles, often used for clustering and classification.
   - `DBSCAN`: A clustering algorithm from `sklearn.cluster` that identifies clusters based on density.

2. **Generating the Data**:


```python
X, _ = make_moons(n_samples=300, noise=0.05, random_state=42)
```

   - `make_moons(n_samples=300, noise=0.05, random_state=42)`:
     - `n_samples=300`: Specifies the number of samples (data points) to generate.
     - `noise=0.05`: Adds random noise to the data points. This helps to make the data more realistic and challenging for clustering algorithms.
     - `random_state=42`: Sets the seed for the random number generator to ensure reproducibility.

   The function returns two values: `X`, the data points, and `_`, which contains the true labels (not used in this case).

3. **Clustering with DBSCAN**:


```python
dbscan = DBSCAN(eps=0.2, min_samples=5)
labels = dbscan.fit_predict(X)
```

   - `DBSCAN(eps=0.2, min_samples=5)`:
     - `eps=0.2`: The maximum distance between two samples for them to be considered as in the same neighborhood. It controls the density criterion for forming clusters.
     - `min_samples=5`: The minimum number of samples required to form a dense region (i.e., a cluster). This parameter helps to determine whether a region is dense enough to be considered a cluster.

   - `dbscan.fit_predict(X)`:
     - `fit_predict(X)`: Fits the DBSCAN model to the data and assigns cluster labels to each data point. The result, `labels`, is an array where each element is the cluster label of the corresponding data point. Outliers are labeled as `-1`.

4. **Plotting the Clusters**:


```python
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='Paired', marker='o')
plt.title("DBSCAN Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

- `plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='Paired', marker='o')`:
     - `X[:, 0]` and `X[:, 1]`: The x and y coordinates of the data points.
     - `c=labels`: Colors the data points based on their cluster labels. Outliers will typically be in a different color.
     - `cmap='Paired'`: Specifies the colormap to use for coloring the clusters.
     - `marker='o'`: Specifies the marker style for the data points (circles in this case).

   - `plt.title("DBSCAN Clustering")`: Sets the title of the plot.
   - `plt.xlabel("Feature 1")`: Labels the x-axis.
   - `plt.ylabel("Feature 2")`: Labels the y-axis.
   - `plt.show()`: Displays the plot.

### Summary
This code generates a synthetic dataset of two interleaving half-circles, applies the DBSCAN clustering algorithm to identify clusters based on density, and then visualizes the clustering results using a scatter plot.
