### Basic PLotting

```python
#import matplotlib
import matplotlib.pyplot as plt
```

```python
#create a simple line plot
 
#sample data
x=[1,2,3,4,5]
y=[10,11,13,14,16]
plt.plot(x,y)
```

## Customizing the matplotlib

```python
plt.plot(x,y)

## add the title and the label
plt.title('Plotting a simple line')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

### customizing our plots

- We can customize our plots by adding labels, titles, and legends. We can also change the colors.
- View further details in the [link ➡️](https://matplotlib.org/stable/plot_types/basic/plot.html#sphx-glr-plot-types-basic-plot-py)

```python
plt.plot(x,y,marker='o',linestyle = '--',color='r')
plt.show()
```

### Changing the figure size

```python
plt.figure(figsize=(8,7))
plt.plot(x,y)
plt.show()
```

### Adding legends


```python
plt.plot(x,y,label='Line 1')
plt.legend()
```

### Annotations


```python
plt.plot(x,y,label='Line 1')
plt.legend()

plt.annotate("Point of Interest",xy=(2,11),xytext=(3,13),
             arrowprops=dict(facecolor='black', shrink=0.05))
```

### create a subplot

```python
fig, axs = plt.subplots(2,2)
axs[0,0].plot(x,y)
axs[0,1].plot(y,x)
axs[1,0].plot(x,x)
axs[1,1].plot(y,y)
plt.show()
```

## 3D plots

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x,y,zs=0,zdir = 'z', label='zs=0,zdir=z')
plt.show()
```

# Seaborn library

```python
import seaborn as sns
import matplotlib.pylab as plt
```

- load data from seaborn library [[link➡️]](https://seaborn.pydata.org/generated/seaborn.load_dataset.html)
- github for data [[GIthub link➡️]](https://github.com/mwaskom/seaborn-data)

```python
# create a simpledataset using tips
tips = sns.load_dataset("tips")
```

```python
sns.scatterplot(data=tips,x="total_bill",y='tip')
```

```python
sns.scatterplot(data=tips,x="total_bill",y='tip',hue='day')
```

- ploting with different styles


```python
sns.set_style('whitegrid')
sns.scatterplot(data=tips,x="total_bill",y='tip',hue='day')
```

```python
# facet grid
g = sns.FacetGrid(tips,col='time')
g.map(sns.scatterplot,"total_bill",'tip')
```

```python
tips.head()
```

```python
# pair plots
sns.pairplot(tips)
plt.show()
```

# HeatMap
- shows co-relation between two variables


```python
# corr= tips.corr
# corr
import pandas as pd
data={
    "Age":[19,20,23,25,23],
    "score":[90,70,60,80,50],
    "study":[4,6,7,5,9]
}
df = pd.DataFrame(data)
df
```

```python
corr = df.corr()
sns.heatmap(corr,annot=True)
plt.show()
```
