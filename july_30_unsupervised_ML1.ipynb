# Clustering Algorithms
## K-Means Clustering

- it involves partitioning the data into k-clusters, minimizing the variabce within the cluster

### Steps
1. Choose the number of clusters
2. Innitialize K centroids randomly
3. Assign each of the points to the nearest cluster
4. Recalculate the centroids and adjust based on the mean of all the data points
5. Repeat step 3 and 4 until a convergence


#### Code Block 1: Importing Libraries

````python
# Implementation
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
````

**`numpy (np)`**: A fundamental package for array computing with Python.
- **`matplotlib.pyplot (plt)`**: A plotting library used for creating visualizations.
- **`sklearn.datasets.make_blobs`**: A function to generate synthetic datasets, which is often used for clustering problems.
- **`sklearn.cluster.KMeans`**: A class implementing the K-Means clustering algorithm.

#### Code Block 2: Data Generation


````python
# Generate some data
X, y = make_blobs(n_samples=500, centers=4, cluster_std=0.80, random_state=42)
X,y
````

**`n_samples=500`**: Specifies the number of data points to generate.
- **`centers=4`**: Specifies the number of centers (clusters) to generate.
- **`cluster_std=0.80`**: The standard deviation of the clusters, which controls how spread out the points are.
- **`random_state=42`**: A seed for the random number generator to ensure reproducibility.

##### Here, `X` contains the generated feature data, and `y` contains the labels indicating the actual cluster each point belongs to.


````python
X, y
````

#### Code Block 3: Applying K-Means Clustering

````python
# Apply the KMeans 
kmeans = KMeans(n_clusters=4)
y_kmeans = kmeans.fit_predict(X)
````

- **`n_clusters=4`**: Specifies the number of clusters the algorithm should identify.
- **`fit_predict(X)`**: This method first fits the model to the data `X` and then predicts the cluster for each data point.

The output `y_kmeans` contains the cluster labels assigned by the K-Means algorithm.
#### Code Block 4: Visualizing the Clusters

````python
# Visualize the clusters
plt.scatter(X[:, 0], X[:, 1], c = y_kmeans, s= 50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s = 100, alpha = 0.75, marker = 'X')
plt.title("KMeans clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
````

The code snippet you provided is for visualizing the clusters formed by a KMeans algorithm using Matplotlib. Here's a breakdown of what each line does:

- `plt.scatter(X[:, 0], X[:, 1], c = y_kmeans, s= 50, cmap='viridis')`: This line creates a scatter plot of the data points. `X[:, 0]` and `X[:, 1]` represent the first and second features of the data. The color of each point is determined by the cluster it belongs to (`y_kmeans`). The `s=50` sets the size of the points, and `cmap='viridis'` applies the 'viridis' colormap.

- `centers = kmeans.cluster_centers_`: This line extracts the coordinates of the cluster centers from the `kmeans` object.

- `plt.scatter(centers[:, 0], centers[:, 1], c='red', s = 100, alpha = 0.75, marker = 'X')`: This line plots the cluster centers on the same scatter plot using red 'X' markers. `s=100` increases the size of the markers, and `alpha = 0.75` makes them semi-transparent.

- `plt.title("KMeans clustering")`: Adds a title to the plot.

- `plt.xlabel("Feature 1")` and `plt.ylabel("Feature 2")`: Label the x and y axes as "Feature 1" and "Feature 2".

- `plt.show()`: Displays the plot.

If you run this code, it will produce a scatter plot showing the clusters with their respective centroids, allowing you to visually assess the clustering result.

# Hierarchical Clustering


Hierarchical clustering is a method of cluster analysis that seeks to build a hierarchy of clusters. It is commonly used for data that does not naturally split into a predetermined number of clusters. There are two main types of hierarchical clustering: **agglomerative** and **divisive**.

### 1. **Agglomerative Hierarchical Clustering (Bottom-Up)**
   - **Process:**
     - **Start with individual points as their own clusters:** Initially, each data point is treated as a single cluster.
     - **Merge the closest pair of clusters:** At each step, the algorithm merges the two closest clusters based on a chosen distance metric (e.g., Euclidean distance).
     - **Repeat until all points are in a single cluster:** This process continues until all data points are merged into a single cluster, forming a hierarchical tree called a **dendrogram**.

     
   - **Distance Metrics:**
     - **Single Linkage (Minimum Linkage):** The distance between two clusters is defined as the shortest distance between any two points in the two clusters.
     - **Complete Linkage (Maximum Linkage):** The distance between two clusters is the maximum distance between any two points in the two clusters.
     - **Average Linkage:** The distance between two clusters is the average distance between all pairs of points in the two clusters.
     - **Ward’s Method:** This method aims to minimize the variance within clusters. It calculates the sum of squared differences within all clusters and merges the clusters that lead to the smallest increase in the sum.
   - **Characteristics:**
     - Often results in a more intuitive hierarchical tree (dendrogram).
     - Requires no predefined number of clusters; you can decide on the number of clusters by cutting the dendrogram at a desired level.

### 2. **Divisive Hierarchical Clustering (Top-Down)**
   - **Process:**
     - **Start with all points in a single cluster:** Initially, all data points are grouped into one large cluster.
     - **Recursively split the clusters:** The algorithm repeatedly divides the clusters into smaller ones until each data point is its own cluster.
     - **Stop when necessary:** This process continues until the desired number of clusters is achieved or until each data point is isolated.
   - **Splitting Methods:**
     - The splitting is often based on a criterion that maximizes the distance between the resulting clusters.
     - Unlike agglomerative methods, divisive clustering tends to be computationally more expensive, especially for large datasets.
   - **Characteristics:**
     - This approach is less commonly used in practice because it is computationally more intensive than agglomerative clustering.
     - However, it can be useful in cases where a top-down perspective on the clustering is needed.

### **Dendrogram**
   - A dendrogram is a tree-like diagram that records the sequences of merges or splits.
   - In agglomerative clustering, a dendrogram illustrates the order in which clusters were merged, with the height of the merge representing the distance or dissimilarity between the merged clusters.
   - By cutting the dendrogram at a particular level, you can decide on the number of clusters you want to retain.

### **Comparison of Hierarchical and KMeans Clustering**
   - **Hierarchical Clustering**:
     - Does not require the number of clusters to be specified in advance.
     - More flexible since it can result in a hierarchy of clusters.
     - Computationally expensive for large datasets.
   - **KMeans Clustering**:
     - Requires the number of clusters to be specified in advance.
     - Faster and more efficient for large datasets.
     - No hierarchy of clusters; you get a flat partitioning.

### **Applications**
   - **Agglomerative Clustering:** Often used in bioinformatics (e.g., gene expression data), market segmentation, and document clustering.
   - **Divisive Clustering:** Less commonly used, but can be applied where a top-down clustering structure is needed, such as in certain types of network analysis.

Hierarchical clustering, particularly the agglomerative approach, is popular due to its simplicity and the interpretability of the dendrogram, which can provide insights into the data structure.

#### Code Block 1: Importing Libraries




````python
from scipy.cluster.hierarchy import linkage, dendrogram
````

- **`scipy.cluster.hierarchy.linkage`**: Computes the hierarchical clustering.
- **`scipy.cluster.hierarchy.dendrogram`**: Generates a dendrogram to visualize the hierarchical clustering.

#### Code Block 2: Data Generation


````python
# Generate the data
X, y = make_blobs(n_samples=50, centers=4, cluster_std=0.5, random_state=42)
````

Similar to the K-Means example, but with fewer samples and tighter clusters.

#### Code Block 3: Performing Hierarchical Clustering

````python
# Perform a hierachical clustering
Z = linkage(X, 'ward')
````

- **`linkage(X, 'ward')`**: Performs hierarchical/agglomerative clustering using the Ward variance minimization algorithm, which minimizes the total within-cluster variance.

The result, `Z`, contains the hierarchical clustering encoded as a linkage matrix.

#### Code Block 4: Plotting the Dendrogram


````python
# Plot the dendrogram
plt.figure(figsize=(10,7))
dendrogram(Z)
plt.title("Hierachical Clustering Dendrogram")
plt.xlabel('sample index')
plt.ylabel('Distance')
plt.show()
````

```
- **`figsize=(10, 7)`**: Specifies the size of the figure.
- **`dendrogram(Z)`**: Plots the hierarchical clustering as a dendrogram.
- **`xlabel('sample index')`**: Labels the x-axis with 'sample index'.
- **`ylabel('Distance')`**: Labels the y-axis with 'Distance', indicating how far apart clusters are.

This dendrogram visually represents the merging process of hierarchical clustering.
