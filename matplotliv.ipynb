# Basic Plotting

```python
x= [1,2,3,4,5]
y=[2,3,5,7,11]
```

```python
plt.plot(x,y)
plt.xlabel("x range")
plt.ylabel("y range")
plt.title("My first plot")
plt.show()
```

```python
# Scatter plot
plt.scatter(x,y)
plt.xlabel('x range')
plt.ylabel('y range')
plt.title('Scatter plot')
plt.show()
```

### bar plot
 - used to compare quantities

```python
Categories = ['A','B','C','D']
Values =[3,4,5,7]
```

```python
plt.bar(Categories,Values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar plot')
plt.show()
```

## Histogram
 - Distribution of the dataset

```python
data = [1,2,3,2,3,5,7,6,5,9,11,15,9]
plt.hist(data,bins=5)
plt.xlabel('Vlaues')
plt.ylabel('Frequency')
plt.title('Simple Histogram')
plt.show()
```

## Outliers
- values that are identified by Box plots
 ### Box plots
  - show the distribution of the data based on 5 summary statistics
  - Minimum value, Max value , First quartile, Median, Third Quartile,

```python
plt.boxplot(data)
plt.title('Box plot')
plt.show()
```
